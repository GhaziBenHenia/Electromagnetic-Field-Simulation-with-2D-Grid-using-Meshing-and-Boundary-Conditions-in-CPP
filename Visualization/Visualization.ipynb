```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load data from CSV file
data = pd.read_csv('../testOutputSamples/output_field.csv')

grid_size = 5  # Adjust based on your grid
electric_field_data = data['Ex'].values.reshape((grid_size, grid_size))
magnetic_field_data = data['Bz'].values.reshape((grid_size, grid_size))

fig, axes = plt.subplots(2, 1, figsize=(8, 12))  # Increased size for higher resolution

# Electric Field Visualization
c1 = axes[0].imshow(electric_field_data, cmap='RdBu', interpolation='bilinear')
axes[0].set_title('Electric Field Distribution')
fig.colorbar(c1, ax=axes[0], label='Electric Field (Ex)')

# Magnetic Field Visualization
c2 = axes[1].imshow(magnetic_field_data, cmap='RdBu', interpolation='bilinear')
axes[1].set_title('Magnetic Field Distribution')
fig.colorbar(c2, ax=axes[1], label='Magnetic Field (Bz)')

plt.tight_layout()
plt.savefig("field_visualization_high_res.png", dpi=300)  # Save with high resolution
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('../testOutputSamples/wave_propagation.csv')

# Get unique time steps to visualize the wave at different stages
time_steps = data['Time'].unique()

# Define grid size based on the number of nodes
num_nodes = len(data[data['Time'] == time_steps[0]])
grid_size = int(np.sqrt(num_nodes))

# Choose a specific time step for visualization (e.g., middle of simulation)
selected_time = time_steps[len(time_steps) // 2]

# Extract data for the selected time step
subset = data[data['Time'] == selected_time]
electric_field_data = subset['ElectricField'].values.reshape((grid_size, grid_size))
magnetic_field_data = subset['MagneticField'].values.reshape((grid_size, grid_size))

fig, axes = plt.subplots(2, 1, figsize=(10, 14))  # Increased size for higher resolution

# Electric Field Visualization
c1 = axes[0].imshow(electric_field_data, cmap='RdBu', interpolation='bilinear')
axes[0].set_title('Electric Field Distribution (Time: {:.2f}s)'.format(selected_time))
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
fig.colorbar(c1, ax=axes[0], label='Electric Field (Ex)')

# Magnetic Field Visualization
c2 = axes[1].imshow(magnetic_field_data, cmap='RdBu', interpolation='bilinear')
axes[1].set_title('Magnetic Field Distribution (Time: {:.2f}s)'.format(selected_time))
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
fig.colorbar(c2, ax=axes[1], label='Magnetic Field (Bz)')

plt.tight_layout()
plt.savefig("field_visualization_high_res.png", dpi=300)  # Save with high resolution
plt.show()
```
